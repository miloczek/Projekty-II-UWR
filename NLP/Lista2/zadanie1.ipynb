```python
# Tworzę struktury przechowujące ilość poszczególnych słów oraz długości zdań

def word_counter(file_path):
    result_dict = {}
    with open(file_path, 'r', encoding='utf-8') as data:
        for line in data:
            words = line.split()
            for word in words:
                if word != ' ':
                    if word in result_dict.keys():
                        result_dict[word] += 1
                    else:
                        result_dict[word] = 1
    return result_dict


def sentence_lengths_counter(file_path):
    result_dict = {}
    for i in range(201):
        result_dict[i] = 0
    with open(file_path, 'r', encoding='utf-8') as data:
        for line in data:
            if len(line) > 1:
                words = line.split()
                result_dict[len(words)] += 1
    return result_dict

def total_words_counter(word_dict):
    result = 0
    for key in word_dict:
        result += word_dict[key]
    return result



orzeszkowa_words_counted = word_counter('./../ruznezecy/dane_pozytywistyczne/uczaca_orzeszkowa.txt')
prus_words_counted = word_counter('./../ruznezecy/dane_pozytywistyczne/uczacy_prus.txt')
sienkiewicz_words_counted = word_counter('./../ruznezecy/dane_pozytywistyczne/uczacy_sienkiewicz.txt')

orzeszkowa_lenghts_sentence_counted = sentence_lengths_counter('./../ruznezecy/dane_pozytywistyczne/uczaca_orzeszkowa.txt')
prus_lenghts_sentence_counted = sentence_lengths_counter('./../ruznezecy/dane_pozytywistyczne/uczacy_prus.txt')
sienkiewicz_lenghts_sentence_counted = sentence_lengths_counter('./../ruznezecy/dane_pozytywistyczne/uczacy_sienkiewicz.txt')

orzeszkowa_total_words = total_words_counter(orzeszkowa_words_counted)
prus_total_words = total_words_counter(prus_words_counted)
sienkiewicz_total_words = total_words_counter(sienkiewicz_words_counted)

orzeszkowa_total_lengths = total_words_counter(orzeszkowa_lenghts_sentence_counted)
prus_total_lengths = total_words_counter(prus_lenghts_sentence_counted)
sienkiewicz_total_lenghts = total_words_counter(sienkiewicz_lenghts_sentence_counted)








```

```python
import math

def nb_checker(list_of_words):
    result = {'orzeszkowa' : 1, 'prus' : 1, 'sienkiewicz' : 1}
    premia = 2.5
    for word in list_of_words:
        if word in orzeszkowa_words_counted.keys():
            result['orzeszkowa'] += orzeszkowa_words_counted[word] / orzeszkowa_total_words
        else:
            result['orzeszkowa'] += 0.1 / orzeszkowa_total_words
        
        if word in prus_words_counted.keys():
            result['prus'] += prus_words_counted[word] / prus_total_words
        else:
            result['prus'] += 0.1 / prus_total_words

        if word in sienkiewicz_words_counted.keys():
            result['sienkiewicz'] += sienkiewicz_words_counted[word] / sienkiewicz_total_words
        else:
            result['sienkiewicz'] += 0.1 / sienkiewicz_total_words

    if len(list_of_words) in orzeszkowa_lenghts_sentence_counted:
        result['orzeszkowa'] += premia * orzeszkowa_lenghts_sentence_counted[len(list_of_words)] / orzeszkowa_total_lengths
    else:
        result['orzeszkowa'] += 0.1 / orzeszkowa_total_lengths

    if len(list_of_words) in prus_lenghts_sentence_counted:
        result['prus'] += premia * prus_lenghts_sentence_counted[len(list_of_words)] / prus_total_lengths
    else:
        result['prus'] += 0.1 / prus_total_lengths

    if len(list_of_words) in sienkiewicz_lenghts_sentence_counted:
        result['sienkiewicz'] += premia * sienkiewicz_lenghts_sentence_counted[len(list_of_words)] / sienkiewicz_total_lenghts
    else:
        result['sienkiewicz'] += 0.1 / sienkiewicz_total_lenghts
    
    return result



def orzeszkowa_validation(file_path):
    sentences, good_sentences, total_sentences = [], 0, 0
    with open(file_path, 'r', encoding='utf-8') as walidacyjna_orzeszkowa:
        for line in walidacyjna_orzeszkowa:
                if len(line) > 1:
                    sentences.append(line)
    for sen in sentences:
        words = sen.split()
        nb_result = nb_checker(words)
        maxi = 0
        for key in nb_result:
            if nb_result[key] > maxi:
                maxi = nb_result[key]
        for key in nb_result:
            if nb_result[key] == maxi:
                if key == 'orzeszkowa':
                    good_sentences += 1
        total_sentences += 1
    return good_sentences / total_sentences

def prus_validation(file_path):
    sentences, good_sentences, total_sentences = [], 0, 0
    with open(file_path, 'r', encoding='utf-8') as walidacyjny_prus:
        for line in walidacyjny_prus:
                if len(line) > 1:
                    sentences.append(line)
    for sen in sentences:
        words = sen.split()
        nb_result = nb_checker(words)
        maxi = 0
        for key in nb_result:
            if nb_result[key] > maxi:
                maxi = nb_result[key]
        for key in nb_result:
            if nb_result[key] == maxi:
                if key == 'prus':
                    good_sentences += 1
        total_sentences += 1
    return good_sentences / total_sentences


def sienkiewicz_validation(file_path):
    sentences, good_sentences, total_sentences = [], 0, 0
    with open(file_path, 'r', encoding='utf-8') as walidacyjny_sienkiewicz:
        for line in walidacyjny_sienkiewicz:
                if len(line) > 1:
                    sentences.append(line)
    for sen in sentences:
        words = sen.split()
        nb_result = nb_checker(words)
        maxi = 0
        for key in nb_result:
            if nb_result[key] > maxi:
                maxi = nb_result[key]
        for key in nb_result:
            if nb_result[key] == maxi:
                if key == 'sienkiewicz':
                    good_sentences += 1
        total_sentences += 1
    return good_sentences / total_sentences

        


print('wynik orzeszkowej:', orzeszkowa_validation('./../ruznezecy/dane_pozytywistyczne/walidacyjna_orzeszkowa.txt'))
print('wynik prusa:', prus_validation('./../ruznezecy/dane_pozytywistyczne/walidacyjny_prus.txt'))
print('wynik sienkiewicza', sienkiewicz_validation('./../ruznezecy/dane_pozytywistyczne/walidacyjny_sienkiewicz.txt'))
```
